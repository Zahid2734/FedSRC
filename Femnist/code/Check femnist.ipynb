```python
import matplotlib.pyplot as plt

def plot_distribution(data):
    plt.hist(data, bins='auto', alpha=0.7, rwidth=0.85)
    plt.grid(axis='y', alpha=0.5)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Distribution Plot')
    plt.show()
```

```python
import pickle
def save_file(file_name,data):
    open_file = open(file_name, "wb")
    pickle.dump(data, open_file)
    open_file.close()

def open_file(file_name):
    open_file = open(file_name, "rb")
    Dataset = pickle.load(open_file)
    open_file.close()
    return Dataset
```

```python
file_name="../data/fedavg femnist shuffle flip noise.pkl"
fedavg=open_file(file_name)
```

```python
import numpy as np
plot_distribution(fedavg[3][9])
plot_distribution(fedavg[5][10])
print('training median', np.median(fedavg[3][9]), 'standard deviation', np.std(fedavg[3][9]))
print('testing median', np.median(fedavg[5][10]), 'standard deviation', np.std(fedavg[5][10]))
```
